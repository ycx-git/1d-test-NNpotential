```python
from IPython import display
import numpy as np
import matplotlib
import torch
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

La=-10
Lb=10
N=3000
en_num=120
x=np.linspace(La, Lb, N+2)
x=x[1:-1]

for i in range(200):
    i=i*60
    file_name=f'./fun_images/V_{La}_{Lb}_{N}_{en_num}/V_diag_{i}.pth'
    V_NN=torch.load(file_name)
    V_NN=V_NN.detach().cpu().numpy()
    plt.cla()  # 清除图片
    plt.xlim((La-2, Lb+2))
    plt.ylim((-100, 300))
    plt.plot(x, V_NN, color='black')  # 绘制曲线
    display.clear_output(wait=True)
    plt.title(f'V_diag_{i}')
    plt.pause(0.01)
plt.show()
    
```
