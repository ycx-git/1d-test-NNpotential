```python
import torch
import numpy as np
a=torch.tensor([1, 2, 3])

print(a.repeat(10,1))

b=np.array([4, 5, 6])
print(a + b)
```

```python
x=np.linspace(0, 1, 50)
y=np.sin(x*2*np.pi)
z=np.cos(x*2*np.pi)
import matplotlib.pyplot as plt
plt.plot(x, z, label='cos',zorder=1)
plt.plot(x, y, label='sin',zorder=2)

plt.legend()
plt.show()
```
