```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun1(x):
    return 2*x**2
```

```python
def fun2(x):
    return x*torch.sin(torch.pi*x/2.5)*10+x**4/25
```

```python
def fun3(x):
    f=x*torch.sin(torch.pi*x)+x**2*2
    return f
```

```python
def fun4(x):
    f=(x-5)**2*(x+5)**2/25
    return f
```

```python
def fun5(x):
    f=3*x*torch.sin(torch.pi*x*2/3)+x**2*2
    return f
```

```python
def fun6(x):
    f=x*torch.sin(torch.pi*x*2/3)+x**2*2
    return f
```

```python
def fun7(x):
    f=(x/2+torch.sin(x*torch.pi/5))**2*8
    return f
```

```python
def fun8(x):
    f=(x*5/6+torch.sin(x*torch.pi*3/8))**2*3
    return f
```

```python
x=torch.linspace(-10, 10, 1000)
f1=fun1(x)
f2=fun2(x)
f3=fun3(x)
f4=fun4(x)
f5=fun5(x)
f6=fun6(x)
f7=fun7(x)
f8=fun8(x)
```
