```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun1(x):
    return -10/x
```

```python
def fun2(x):
    return -10*torch.sin(x)/x**2
```

```python
def fun3(x):
    f=-10*torch.sin(torch.pi*x/20)**2/x-10/x
    return f
```

```python
def fun4(x):
    f=-18/x-10*torch.sin(torch.pi*x/12)/x
    return f
```

```python
def fun5(x):
    f=-10/x-8*(torch.sin(torch.pi*x/10))**2*torch.exp(-x/15)
    return f
```

```python
def fun6(x):
    f=-10/x-10*torch.sin(torch.pi*x/24)/x
    return f
```

```python
def fun7(x):
    f=-10/x-10*torch.sin(torch.pi*x/24)/x+x/20
    return f
```

```python
def fun8(x):
    f=2*x**2
    return f
```
