```python
import torch
import numpy as np
```

```python
def calc_next(phi_1,phi_0,E,d):
    phi_2=(12-10*E*d**2)*phi_1/(6+E*d**2)-phi_0
    return phi_2
```

```python
def iter(pred_energy,phi_0,phi_1,n,step,lr=0.01):

    phi_n_mines_1=phi_0.clone()
    phi_n=phi_1.clone()
    for i in range(n-1):
        phi_n_plus_1 = calc_next(phi_n,phi_n_mines_1,pred_energy,step)
        phi_n_mines_1 = phi_n
        phi_n = phi_n_plus_1
    loss=phi_n**2
    print('loss:',loss)
    loss.backward()
    pred_energy.data=pred_energy.data-lr*pred_energy.grad.data
    pred_energy.grad.zero_()
    return loss.item()

```

```python
dtype=torch.float32
pred_energy = torch.tensor(0.1,dtype=dtype,requires_grad=True)
const_of_slope=1.0
n=100
step=1/n

phi_0 = torch.tensor(0.0,dtype=dtype)
phi_0_slope = torch.tensor(const_of_slope,dtype=dtype)
phi_1=phi_0+phi_0_slope*step
lr=0.1

for i in range(3000):
    print(pred_energy)
    iter(pred_energy,phi_0,phi_1,n,step,lr)

print(pred_energy)
```
