```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun1(x):
    return -10/x
```

```python
def fun2(x):
    return -10*torch.sin(x)/x**2
```

```python
def fun3(x):
    f=-10*torch.sin(torch.pi*x/20)**2/x-10/x
    return f
```

```python
def fun4(x):
    f=-18/x-10*torch.sin(torch.pi*x/12)/x
    return f
```

```python
def fun5(x):
    f=-10/x-8*(torch.sin(torch.pi*x/10))**2*torch.exp(-x/15)
    return f
```

```python
def fun6(x):
    f=-10/x-10*torch.sin(torch.pi*x/24)/x
    return f
```

```python
def fun7(x):
    f=-10/x-10*torch.sin(torch.pi*x/24)/x+x/20
    return f
```

```python
def fun8(x):
    f=2*x**2
    return f
```

```python
def fun9(x):   
    f=2*x**2+3*torch.sin(torch.pi*x/4.5)*x
    return f
```

```python
def fun10(x):   
    f=2*x**2+3*torch.sin(torch.pi*x/2.4)*x*(1-torch.exp(-x/6))
    return f
```

```python
def fun11(x):   
    f=x**3/20+x*20
    return f
```

```python
def fun12(x):   
    f=x**3/20+torch.sin(torch.pi*x/10)*(20-x)*x+x**2/2
    return f
```

```python
def fun13(x):   
    f=x**3/20+torch.sin(torch.pi*x/10)*(20-x)*x+x**2+x*10
    return f
```

```python
def fun14(x):   
    f=-10/x+x
    return f
```

```python
def fun15(x):   
    f=-10/x+x*10
    return f
```
