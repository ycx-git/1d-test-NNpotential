```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun1(x):
    return -10/x
```

```python
def fun2(x):
    return -10*torch.sin(x)/x**2
```

```python
def fun3(x):
    f=-10*torch.sin(torch.pi*x/20)**2/x-10/x
    return f
```

```python
def fun4(x):
    f=-18/x-10*torch.sin(torch.pi*x/12)/x
    return f
```

```python
def fun5(x):
    f=-10/x-8*(torch.sin(torch.pi*x/10))**2*torch.exp(-x/15)
    return f
```

```python
def fun6(x):
    f=-10/x-10*torch.sin(torch.pi*x/24)/x
    return f
```

```python
def fun7(x):
    f=-10/x-10*torch.sin(torch.pi*x/24)/x+x/20
    return f
```

```python
N=1000
x=torch.linspace(0, 100, N)
# x=x[N//10:-1]
x=x[20:-1]
f1=fun1(x)
f2=fun2(x)
f3=fun3(x)
f4=fun4(x)
f5=fun5(x)
f6=fun6(x)
f7=fun7(x)
```

```python
dirivative=torch.autograd.functional.jacobian(fun7, x)
dirivative=torch.diag(dirivative)
plt.plot(x,torch.zeros_like(x), label='zero')
plt.plot(x,dirivative)
plt.title('Jacobian of fun7')
plt.legend()
plt.show()
```
